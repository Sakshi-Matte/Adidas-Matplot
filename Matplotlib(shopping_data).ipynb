```python
months=['jan','feb','mar','apr','may','jun','july','aug','sep']
profit=[500,700,-100,-150,300,450,550,400,430]
```

```python
plt.bar(months,profit)
plt.xlabel('months')
plt.ylabel('profit in cr')
plt.xticks(rotation=45);
# draw line which is paralell to x ---y intercept
plt.axhline(y=0,color='red')
#plt.axvline(x=0,color='red')
```

```python
#aim to have different color for profit and 
# mar n apr in black rest in green
profit=[500,700,-100,-150,300,450,550,400,430]
col_list=[]
for i in profit:
    if i<0 :
        col_list.append('black')
    else:
        col_list.append('green')
```

```python
col_list
```

```python
plt.bar(months,profit,color=col_list)
plt.xlabel('months')
plt.ylabel('profit in cr')
plt.xticks(rotation=45);
# draw line which is paralell to x ---y intercept
plt.axhline(y=0,color='red')
#plt.axvline(x=0,color='red')
```
